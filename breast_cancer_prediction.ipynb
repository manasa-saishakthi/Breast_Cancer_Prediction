```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
def models(x_train,y_train):
  from sklearn.linear_model import LogisticRegression
  log=LogisticRegression(random_state=0)
  log.fit(x_train,y_train)

  from sklearn.tree import DecisionTreeClassifier
  tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
  tree.fit(x_train,y_train)

  from sklearn.ensemble import RandomForestClassifier
  forest=RandomForestClassifier(criterion='entropy',random_state=0)
  forest.fit(x_train,y_train)

  print('[0]Logistic Regression Training Accuracy:',log.score(x_train,y_train))
  print('[0]Decision Tree Classifier Training Accuracy:',tree.score(x_train,y_train))
  print('[0]Random Forest Classifier Training Accuracy:',forest.score(x_train,y_train))

  return log,tree,forest
```

```python
model=models(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('model',i)
    cm=confusion_matrix(y_test,model[0].predict(x_test))

    TP=cm[0][0]
    TN=cm[1][1]
    FN=cm[1][0]
    FP=cm[0][1]
    print(cm)
    print('Testing Accuracy=', (TP + TN)/ (TP + TN + FN + FP))
    print()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
  print('model',i)
  print(classification_report(y_test,model[i].predict(x_test)))
  print(accuracy_score(y_test,model[i].predict(x_test)))
  print()
```

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of neural network

model = keras.Sequential([
                   keras.layers.Flatten(input_shape=(30,)),
                   keras.layers.Dense(20,activation='relu'),
                   keras.layers.Dense(2,activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the neural network
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'], loc='lower right')
```

```python
loss,accuracy = model.evaluate(x_test_std,y_test)
print(accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
clf = KNeighborsClassifier(n_neighbors = 5)
clf.fit(x_train,y_train)
```

```python
training_acc=clf.score(x_train,y_train)
test_acc=clf.score(x_test,y_test)
```

```python
#these are model's parameters
clf.get_params()
```

```python
#define parameters we are going to iterate through

n_neighbors = [1,5,10,20,50,100]

test_accuracy = {}
train_accuracy={}

for n in n_neighbors:
   clf = KNeighborsClassifier(n_neighbors = n)
   clf.fit(x_train,y_train)
   test_acc = clf.score(x_test,y_test)
   train_acc=clf.score(x_train,y_train)
   test_accuracy[n] = test_acc
   train_accuracy[n] = train_acc
```

```python
for k,v in test_accuracy.items():
    print("test accuracy for {} n_neighbors is {} %".format(k, round(v*100,3)))
```

```python
for k,v in train_accuracy.items():
    print("training accuracy for {} n_neighbors is {} %".format(k, round(v*100,3)))
```
